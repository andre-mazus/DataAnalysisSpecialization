Body mass index calculator

Write a function that calculates the BMI entered by the user, that is, whoever runs it will have to enter this data. You can get more information about its calculation at: ðŸ”— Body Mass Index What is BMI and how is it calculated.

The function must classify the result into its respective categories.

Tip: Try validating the data beforehand, so that it sends a warning message if the data entered by the user is not in the appropriate format or does not take reasonable values.

```python
def BMI_calculator(weight, hight):
    BMI = weight/(hight**2)
    if BMI < 18.5:
        status = "Underweight"
    elif BMI >= 18.5 and BMI <= 24.9:
        status = "Normal"
    elif BMI >= 25 and BMI <= 29.9:
        status = "Overweight"
    else: 
        status = "Obesity"
    print("BMI is:", status)
```

```python
def BMI_calculator_starter():
    weight = (input("Enter the weight in kg (e.g. 70): "))
    try:
        weight = float(weight)
    except:
        print("Please, enter a valid weight!")
        return BMI_calculator_starter()
        
        
    hight = (input("Enter the hight in mt (e.g. 1.70): "))
    try:
        hight = float(hight)
    except:
        print("Please, enter a valid hight!")
        return BMI_calculator_starter()
        
    BMI_calculator(weight, hight)
```

The values â€‹â€‹and keys in the original dictionary are unique; if this is not the case, the function should print a warning message, along with a list of the values â€‹â€‹associated with the repeated key.

```python
meu_dicionario = {'x':'apple', 'y': 'banana', 'z':'banana', 'a':'banana', 'b':'apple'}
```

```python
def dicionario_reverso(dicionario):

    keys = list(dicionario.keys())
    values = list(dicionario.values())
    novo_dicionario = {}

    for i in range(len(values)):

        if values[i] in novo_dicionario:
            new_list = list(novo_dicionario[values[i]])
            new_list.append(keys[i])
            novo_dicionario.update({values[i]:new_list})
            print('Value already exist', new_list)
        else:
            novo_dicionario.update({values[i]:keys[i]})
        
        
    print("OLD:",dicionario)
    print("NEW:",novo_dicionario)
```

```python
def flat_list(my_list):
    result = []
    for item in my_list:
        if isinstance(item, (list, tuple)):
            result.extend(flat_list(item))
        else:
            result.append(item)
    return result
```

```python
def mixed_fraction_to_float(item):
    whole, frac = item.split('-')
    num, denom = frac.split('/')
    return float(whole) + float(num) / float(denom)
```

```python
def float_numbers(first_list):
    new_list = flat_list(first_list)
    number_list = []
    not_number_list = []
    
    for item in new_list:
        try:
            item = float(item)
            number_list.append(item)
        except:
            try:
                item = (w2n.word_to_num(item))
                item = float(item)
                number_list.append(item)
            except:
                try:
                    item = mixed_fraction_to_float(item)
                    item = float(item)
                    number_list.append(item)
                except:
                    not_number_list.append(item)
                        
    return number_list, not_number_list
```

At this point, the client has detected a problem with the passwords used by their employees. Asdf1234, birthdays or similar. To solve this, they have commissioned us to create a Python function that generates more secure passwords. The function must depend on the following parameters:

length (int): Password length
capital letters (bool = True): Whether capital letters should appear
lowercase (bool = True): Whether lowercase letters should appear
numbers (bool = True): Whether numbers should appear
signs (bool = False): If special characters should appear (,-$? or similar)
So, if we execute the function as follows:

create_password(10, True, True, True, True)

We should obtain an output (which we must be able to save) like this:

9Er,5Vn8P$

Make sure that all criteria are met, and that these passwords are truly random.

(EXTRA) Explore how we could make the function automatically copy the password to the computer's clipboard (as if we had selected it and done ctrl+copy).

```python
pip install pyperclip
```

```python
import random
import string
import pyperclip
```

```python
def create_password(length, capital, lowercase, numbers, signs):
    characters =''

    if capital is True:
        capital_letter = string.ascii_letters
        capital_letter = capital_letter.upper()
        characters = characters+capital_letter

    if lowercase is True:
        lower_letter = string.ascii_letters
        lower_letter = lower_letter.lower()
        characters = characters+lower_letter

    if numbers is True:
        numbers = string.digits
        characters = characters+numbers

    if signs is True:
        signs = string.punctuation
        characters = characters+signs

    if characters == '':
        password=''
        return("Imposible to create a password")
    else:
        password = ''.join(random.choices(characters, k=length))
    
    pyperclip.copy(password)
    print("Password has been copied to the clipboard.")
    return password
```

```python
password = create_password(10, True, True, True, True)
print("Password:", password)
```
