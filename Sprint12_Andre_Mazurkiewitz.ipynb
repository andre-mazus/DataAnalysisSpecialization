### Exercise 1

Query the public JSONPlaceholder API using the GET method to obtain:

* List of posts (/posts)
* User list (/users)
* Task list (/all)

```python
url_list = ['https://jsonplaceholder.typicode.com/posts', 'https://jsonplaceholder.typicode.com/users', 'https://jsonplaceholder.typicode.com/todos']
my_dict = {}
for url in url_list:
    response = requests.get(url)
    data = response.json()
    name= url.split("/")[-1]
    my_dict[name] = data
```

```python
for item in my_dict:
    print(item.upper())
    print(json.dumps(my_dict[item], indent = 3))
    print('\n')
```

### Exercise 4

Make a POST request to create a new dummy post. Include the title, message body, and a userId.

* Show the JSON response.
* Show status code

```python
url = 'https://jsonplaceholder.typicode.com/posts'
payload = {
    'title':'title1',
    'body': 'body1',
    'userId': 1,
}
response = requests.post(url, json=payload)
print("Status code:", response.status_code)
data= response.json()
print(data)
```

```python
url = 'https://jsonplaceholder.typicode.com/posts/10'
payload = {
    'title':'title1',
    'body': 'body1',
    'userId': 1,
}
response = requests.patch(url, json=payload)
print("Status code:", response.status_code)
data= response.json()
print(data)
```

### Exercise 1

Use the Open Data BCN API to:

* Search for a dataset of your interest using package_search and package_show.
* From the results obtained, select one that has resources available in CSV or JSON.
* With the resource_id of the selected resource, perform a query using datastore_search to:
    Retrieve at least 100 records.
* Convert the results to a DataFrame.
* Save the DataFrame to a .csv file.

```python
# HELP: https://opendata-ajuntament.barcelona.cat/data/api/3/action/help_show?name=package_search
# HELP: https://opendata-ajuntament.barcelona.cat/data/api/3/action/help_show?name=package_show
```

```python
def package_search(word):
    
    
    # 1-PACKAGE SEARCH
    # Search for the given word, if it is found return the code of the package, if it is not found return not found
    url = "https://opendata-ajuntament.barcelona.cat/data/api/action/package_search"

    params = {
        "q": word
    }
    
    response = requests.get(url, params=params)
    data = response.json()
    
    if data['success']==True and data['result']['count']>0:
        print("Result found in the API")
        code = (data['result']['results'][0]['code'])
    else: 
        return('Not found')



    # 2- PACKAGE SHOW
    #with the code, look and grab the resource IDs of the packages with JSON or CVC extension available and store it in a dic.
    print("Looking for the resource...")
    url1 = "https://opendata-ajuntament.barcelona.cat/data/api/action/package_show"

    params1 = {
        "id": code
    }
    response1 = requests.get(url1, params=params1)
    data1 = response1.json()
    data1 = (data1['result']['resources'])
    ids_extension={}
    for item in data1:
        if item['format'] == 'CSV' or item['format'] == 'JSON':
            ids_extension[item['id']] = item['format']

            
    # 3- DATASTORE SEARCH
    # with the dic of IDs, get the first one that has data available and retrive the data

    url3 = 'https://opendata-ajuntament.barcelona.cat/data/api/action/datastore_search'
    
    for item in ids_extension:

        params3 = {
            "resource_id": item,
            "limit": 10000,
        }
        
        response3 = requests.get(url3, params=params3)
        data3 = response3.json()
        if data3['success'] == True:
            data3 = data3['result']['records']
            print('Resource found.')
            break

            
    # 4- CONVERT TO DATAFRAME
    # with the data obteined before, create a dataframe with it and export to a csv.
    print('Creating DataFrame...')
    df = pd.DataFrame(data3)
    if '_id' or 'ID' or 'id' in df.columns:
        df.set_index('_id', inplace=True)
    #df.to_csv('api_df_csv.csv')
    print('DataFrame exported to a .csv')
    return df



        
    
    
```
